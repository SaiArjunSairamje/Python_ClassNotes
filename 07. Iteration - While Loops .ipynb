```python
#using loops:

#initialization:
number = 1

while number <= 5:

  #perform task
  print(number)

  #here we missed the updation step - and loop will run forever - infinite loop
```

```python
#using loops:

#initialization:
number = 1

while number <= 5:

  #perform task
  print(number)

  #updation step
  number = number + 1


print("AJ")
```

```python
#using loops:

#initialization:
number = 1
count = 0

while number <= 5:

  #perform task
  print(number)

  count += 1
  #updation step
  number = number + 1


print("count = ", count)
```

```python
'''
1.Take number of test_cases (t) as input from the user
2. take t number of inputs from the user
3. print the multiplication table for all these inputs till their 5th multiple

ex:
t = 3
1 - mul table of 1
2 - mul table of 2
3 - mul table of 3

'''
```

```python
#nested loops- a loop within a loop

t = int(input())

i = 1

while i<=t:

  #task - to take input from user and print multiplicaiton table of it
  n = int(input())

  j=1
  #print the multiplication table of n
  while j<=5:
    #print jth multiple of n
    print(n*j, end = " ")
    j += 1

  i = i+1
```

```python
#nested loops- a loop within a loop

t = int(input("enter the number of testcases: "))

i = 1

while i<=t:

  #task - to take input from user and print multiplicaiton table of it
  n = int(input("enter the number you want the table for: "))

  j=1
  #print the multiplication table of n
  print("The multiplication table of ", n, "is: ")

  while j<=5:
    #print jth multiple of n
    print(n*j, end = " ")
    j += 1

  print()
  i = i+1
```
